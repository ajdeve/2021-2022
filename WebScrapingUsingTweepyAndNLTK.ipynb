```python
#import config.py that lists out twitter developer api keys to secure a connection between twitter and python app
from config import *
import tweepy as tw
import pandas as pd
```

```python
#Authentification process for twitter 
auth = tw.OAuthHandler(TWITTER_CONSUMER_KEY, TWITTER_CONSUMER_SECRET)
auth.set_access_token(TWITTER_ACCESS_TOKEN, TWITTER_ACCESS_TOKEN_SECRET)
api = tw.API(auth, wait_on_rate_limit=True)
```

```python
#Setting search keywords
search_words = "squidgame+korea -filter:retweets"
date_since = "2021-10-01"
```

```python
#Print tweets that fit my search criteria
tweets = tw.Cursor(api.search_tweets, tweet_mode='extended', q=search_words, lang="en", since=date_since).items(50)

for tweet in tweets:
    print(tweet.full_text)
```

```python
#Turned tweet data into tabular data format using pandas dataframe 
tweet_dataset = pd.DataFrame(columns=['Tweet Id', 'Tweet Date', 'Follower Count','Retweets', 'Location', 'Tweet Text'])

for tweet in tw.Cursor(api.search_tweets, tweet_mode='extended', q=search_words, lang="en").items(50):
        appending_dataframe = pd.DataFrame([[tweet.id, tweet.created_at, tweet.user.followers_count, tweet.retweet_count, tweet.user.location, tweet.full_text.encode('utf-8')]], columns=['Tweet Id', 'Tweet Date', 'Follower Count', 'Retweets','Location','Tweet Text'])
        tweet_dataset = tweet_dataset.append(appending_dataframe)
```

```python
#Reindexing the dataframe
tweet_dataset = tweet_dataset.reset_index(drop=True)
tweet_dataset.head()
```

```python
#Importing emoji and nltk for data cleaning process
import emoji
import nltk
import re
nltk.download('words')
words = set(nltk.corpus.words.words())
```

```python
#Converting bytes data to string data
tweet_dataset["Tweet Text"] = tweet_dataset["Tweet Text"].str.decode("utf-8")
```

```python
#Data cleaning and creating a new column with cleaned textual data
clean_tweets = []
for key, tweet in tweet_dataset["Tweet Text"].iteritems():
    tweet = tweet.lower()
    tweet = re.sub("@[A-Za-z0-9_]+","", tweet)
    tweet = re.sub("#[A-Za-z0-9_]+","", tweet)
    tweet = re.sub(r"http\S+", "", tweet)
    tweet = re.sub(r"www.\S+", "", tweet)
    tweet = re.sub('[()!?]', ' ', tweet)
    tweet = re.sub('\[.*?\]',' ', tweet)
    tweet = re.sub("[^a-z0-9]"," ", tweet)
    tweet = ''.join(c for c in tweet if c not in emoji.UNICODE_EMOJI) #Remove Emojis
    tweet = tweet.replace("#", "").replace("_", " ") #Remove hashtag sign but keep the text
    tweet = " ".join(w for w in nltk.wordpunct_tokenize(tweet) \
         if w.lower() in words or not w.isalpha())
    clean_tweets.append(tweet)
    
tweet_dataset["Clean Tweet"] = clean_tweets
```

```python
#Checking cleaned data
tweet_dataset["Clean Tweet"].head()
```

```python
# Import TextBlob for text based sentimental analysis
from textblob import TextBlob
```

```python
#Perform sentimental analysis using TextBlob
def sentiment_func(tweet):
    try:
        return TextBlob(tweet).sentiment
    except:
        return None
```

```python
tweet_dataset["Sentiment"] = tweet_dataset["Clean Tweet"].apply(sentiment_func)
tweet_dataset["Sentiment"].head()
```

```python
#Divide the sentimental analysis figures into two columns, polarity and subjectivity 
tweet_dataset["Sentiment"][0][0]
tweet_dataset['Polarity'] = tweet_dataset["Sentiment"].apply(lambda x: x[0])
tweet_dataset['Subjectivity'] = tweet_dataset["Sentiment"].apply(lambda x: x[1])
tweet_dataset[['Clean Tweet','Sentiment','Polarity','Subjectivity']].head()
```

```python
#Assign words to polarity column
sentiment_word = []
for key, value in tweet_dataset["Polarity"].iteritems():
    if value > 0:
         value ='positive'
    elif value == 0:
         value ='neutral'
    else:
         value ='negative'
    sentiment_word.append(value)
tweet_dataset["Sentiment Word"] = sentiment_word
tweet_dataset[['Clean Tweet','Sentiment Word']]
```

```python
#Calculate the number of positive, negative and neutral tweets 
tweets = tweet_dataset['Sentiment Word'].count()
positive = tweet_dataset[tweet_dataset['Sentiment Word'] == 'positive'].shape[0]
negative = tweet_dataset[tweet_dataset['Sentiment Word'] == 'negative'].shape[0]
neutral = tweet_dataset[tweet_dataset['Sentiment Word'] == 'neutral'].shape[0]
```

```python
#Calculate the percentage of positive, negative and neutral words based on 50 recent searchs and display 5 tweets per each sentiment
pd.set_option('display.max_colwidth', None)
print("Among the recent 50 tweets on search term, 'Squid Game'")
print("Positive tweets percentage: {} %".format(100*positive/tweets))
print(tweet_dataset.loc[tweet_dataset['Sentiment Word'] == 'positive',['Tweet Text']].head())
print("----------------------------------------------------------------")
print("Negative tweets percentage: {} %".format(100*negative/tweets))
print(tweet_dataset.loc[tweet_dataset['Sentiment Word'] == 'negative',['Tweet Text']].head())
print("----------------------------------------------------------------")
print("Neutral tweets percentage: {} % ".format(100*neutral/tweets))
print(tweet_dataset.loc[tweet_dataset['Sentiment Word'] == 'neutral',['Tweet Text']].head())
```
