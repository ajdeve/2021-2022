```python
import csv
import requests
from bs4 import BeautifulSoup
from IPython.display import HTML
import pandas as pd
import numpy as np
```

```python
#Scrap the list of living centenarians
url = 'https://en.wikipedia.org/wiki/List_of_living_centenarians'
headers = {'User-Agent': "Chrome/54.0.2840.90"}
response = requests.get(url, headers=headers, verify=False)
html = response.content 
```

```python
HTML('<iframe src=https://en.wikipedia.org/wiki/List_of_living_centenarians width=900 height=350></iframe>')
```

```python
#lxml is the most feature-rich and easy-to-use library for processing XML and HTML in the Python language.
soup = BeautifulSoup(html, 'lxml')
tables = soup.find_all('table', {'class':'wikitable sortable'})
```

```python
tables2 = tables[0]
```

```python
tables2
```

```python
#Check data is correctly brought up
row = tables2.find_all('tr')[1]
print(tables2.find_all('tr')[1])
cell = row.find_all('td')
print(row.find_all('td'))
print("----")
print(cell[1].find(text=True))
print(cell[4].find(text=True))
print(cell[5].find(text=True))
```

```python
#Scrap data into lists
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
for row in tables2.find_all('tr'):
    cells=row.find_all('td')
    if len(cells)>1:
        A.append(cells[0].text.strip())
        B.append(cells[1].text.strip()) 
        C.append(cells[2].text.strip())
        D.append(cells[3].text.strip())
        E.append(cells[4].text.strip())
        F.append(cells[5].text.strip())
```

```python
#Lists to pandas dataframe 
df=pd.DataFrame(A,columns=['Name'])
df['Gender']=B
df['Birth Date']=C
df['Age']=D
df['Nationality']=E
df['Notability']=F
df.head(50)
```

```python
#Clean up the data
import re

lines = []
for i in F:
    lines.append(re.sub("\[[0-9]+\]", '', i.strip()))
```

```python
df['Notability']= lines
```

```python
df.index = np.arange(1, len(df) + 1)
df.head(50)
```
